```python
import os
import random
import matplotlib.pyplot as plt
import cv2

# Paths
damaged_path = "../data/raw/AI_for_Art_Restoration_2/paired_dataset_art/damaged"
undamaged_path = "../data/raw/AI_for_Art_Restoration_2/paired_dataset_art/undamaged"

heritage_path = "../data/raw/Cultural Product Design Dataset"

# Helper to show random samples
def show_random_images(folder, n=5, title="Sample Images"):
    files = os.listdir(folder)
    sample_files = random.sample(files, min(n, len(files)))
    
    plt.figure(figsize=(15,3))
    for i, fname in enumerate(sample_files):
        img = cv2.imread(os.path.join(folder, fname))
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        plt.subplot(1, n, i+1)
        plt.imshow(img)
        plt.axis("off")
        plt.title(fname[:12])
    plt.suptitle(title)
    plt.show()

# 1. Dataset stats
print("Damaged images:", len(os.listdir(damaged_path)))
print("Undamaged images:", len(os.listdir(undamaged_path)))
print("Heritage dataset images:", len(os.listdir(heritage_path)))

# 2. Show random samples
show_random_images(damaged_path, 5, "Damaged Artifacts")
show_random_images(undamaged_path, 5, "Undamaged Artifacts")
show_random_images(heritage_path, 5, "Heritage Items")
```
